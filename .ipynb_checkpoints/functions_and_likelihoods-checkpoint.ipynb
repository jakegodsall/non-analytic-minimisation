```python
def plot_models(model, xlims=(-10, 10), ylims=(-10, 10)):
    model_obj = Model(model)
    sample = model_obj.generate_random_sample(42, 2, 2, 100)
    plots = model_obj.plot(sample, xlims, ylims)
```

```python
def model(params, x=None):
    a, b = params
    return ((a + 2*b - 7)**2 + (2*a + b - 5)**2)

plot_models(model)
```
